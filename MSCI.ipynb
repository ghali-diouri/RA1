```python
# Tickers list obtained from the 'index.ipynb' notebook
filename = 'stocks.pkl'
with open(filename, 'rb') as f : 
   l = pickle.load(f)

l ;
```

```python


df = data [data['ISSUER_TICKER'].isin(l)]
df = df.dropna(subset = ['ISSUER_TICKER'])

characteristic = 'ENVIRONMENTAL_PILLAR_SCORE'

df = df[['ISSUER_NAME','ISSUER_TICKER',characteristic,'AS_OF_DATE']]
df['AS_OF_DATE'] = df['AS_OF_DATE'].apply(lambda x : pd.Timestamp(str(x)))

df_scaled = pd.DataFrame([])
for x in list(df['AS_OF_DATE'].unique()) : 
    table = df [df['AS_OF_DATE'] == x]
    table ['demeaned'] = scale(table[characteristic], with_std = False)
    table ['demeaned_rescaled'] = minmax_scale(table ['demeaned'], (-0.5, 0.5))
    df_scaled = pd.concat((df_scaled, table))
```
