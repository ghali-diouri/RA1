```python
def rescale(characteristic, data) : 
    df = data [data['ISSUER_TICKER'].isin(l)]
    df = df.dropna(subset = ['ISSUER_TICKER'])
    df = df[['ISSUER_NAME','ISSUER_TICKER',characteristic,'AS_OF_DATE']]
    df.loc[:,'AS_OF_DATE'] = df['AS_OF_DATE'].apply(lambda x : pd.Timestamp(str(x)))

    df_scaled = pd.DataFrame([])
    for x in list(df['AS_OF_DATE'].unique()) : 
        table = df [df['AS_OF_DATE'] == x]
        table.loc[:,'demeaned'] = scale(table[characteristic], with_std = False)
        table.loc[:,'demeaned_rescaled'] = minmax_scale(table ['demeaned'], (-0.5, 0.5))
        df_scaled = pd.concat((df_scaled, table))
    return df_scaled
```

```python
msci = pd.DataFrame([])
for df in list_data :
    msci = pd.concat((msci,rescale(characteristic = characteristic, data = df )))

msci['AS_OF_DATE'] = msci['AS_OF_DATE'].apply(lambda x : x + MonthEnd())

filename = 'msci.pkl'
with open(filename, 'wb') as f : 
   pickle.dump(msci,f)
```
