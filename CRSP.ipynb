```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
data = pd.read_csv('data.csv')
```

```python
data['marketcap'] = (data['PRC'] * data['SHROUT'])
data
```

# Removing negative market capitalizations

```python
data['date1'] = data['date'].apply(lambda x : pd.Timestamp(str(x)))
data['RET1'] = data['RET'].replace('C',0).astype(float)
```

```python
(data['PRC'] < 0).value_counts(normalize= True)
```

```python
lst = pd.DataFrame(data.groupby('COMNAM').nth(0)['marketcap'] < 0)
out = list(lst[lst.marketcap].index)
out
```

```python
data[data['date']==20090930] #.groupby('COMNAM').agg()#['marketcap'] < 0
```

```python
df = data[~data['COMNAM'].isin(out)] #remove negative marketcaps
df
```

# Remove microcap stocks

```python
from numpy import isin


microcaps = pd.DataFrame(df.groupby('COMNAM').nth(0))
microcaps = list(microcaps[microcaps.marketcap < microcaps[microcaps['EXCHCD']==1].marketcap.quantile(0.2)].index)
#px.histogram(df.groupby('COMNAM').nth(0), 'marketcap', nbins = 50)
```

```python
df = df[~df['COMNAM'].isin(microcaps)]
df
```

# Remove penny stocks

```python
penny = pd.DataFrame(df.groupby('COMNAM').nth(0)['PRC'] < 5)
pennystocks = list(penny[penny.PRC].index)
```

```python
df = df[~df['COMNAM'].isin(pennystocks)]
```

# Let's now take a look at the dataset
